Use ** df3 ** para replicar os seguintes gráficos.

** Crie um histograma da coluna 'a'. **

```python
df3['a'].plot.hist()
```

** Estes plots estão ok, mas elas não parecem muito profissionais. Use folhas de estilo para definir o estilo em 'ggplot' e refazer o histograma de cima. Veja também como adicionar mais barras.

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df3['a'].hist(alpha=(.5), bins = 25)
```

** Crie um boxplot comparando as colunas a e b. **

```python
dfab = df3[['a','b']]
dfab.plot.box()
```

** Criar um plot "kde" da coluna 'd' **

```python
df3['d'].plot.kde()
```

** Descubra como aumentar a largura de linha e fazer o linestyle pontilhado. **

```python
df3['d'].plot.kde(linestyle='dashed', linewidth=4)
```

** Crie um gráfico de área de todas as colunas para apenas as linhas até 30. (sugestão: use .ix). **

```python
ax = df3.plot.area(alpha=0.3, figsize=(6,4))
ax.set_xlim(0, 30)
ax.set_ylim(0, 2.8)
```

## Desafio!
** Observe como a legenda em nossa figura anterior sobrepôs alguns dos diagramas reais. Você pode descobrir como exibir a legenda fora do gráfico, conforme mostrado abaixo? **

```python
ax = df3.plot.area(alpha=0.3, figsize=(6,4))
plt.legend(bbox_to_anchor=(1.08, 0.5), loc='center', borderaxespad=0.)
ax.set_xlim(0, 30)
ax.set_ylim(0, 2.8)
```
